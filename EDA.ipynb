# Exploratory Data Analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

eda_df = pd.read_csv("training_data.csv")

# üöÄ General statistics summary
print("üîç General Statistics Summary:")
print(eda_df.describe())

eda_numeric = eda_df.select_dtypes(include=['number'])

plt.figure(figsize=(7, 8))
sns.heatmap(eda_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("üîç Feature Correlation Heatmap")
plt.show()

eda_numeric.hist(figsize=(8,12), bins=30, edgecolor='black')
plt.suptitle("üìä Feature Distributions", fontsize=16)
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x=eda_df["danceability"], y=eda_df["energy"], alpha=0.6, edgecolor=None)
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("üìä Danceability vs. Energy")
plt.show()
```
